```python
import matplotlib.pyplot as plt
import numpy as np
```

# Line plot

```python
x = [1,2,3,4,5]
math = [5,10,15,20,15]
eng = [4,8,12,16,20]

plt.plot(x,math,label="math",color="red",marker="o",linestyle="-",linewidth=2,markersize=8,markerfacecolor="blue",markeredgecolor="black")
plt.plot(x,eng,label="english",color="blue",marker="s",linestyle="--",linewidth=2,markersize=8)

plt.xlabel("students")
plt.ylabel("no of hours studied")
plt.title("Study hours vs students",fontdict={"fontsize":16,"fontweight":10,"fontname":"arial","color":"green"})
plt.legend(loc="upper left")
plt.show()

plt.xticks([1,2,5])
plt.yticks([0,5,10,15,20])

plt.plot(x,math,"ro-",x,eng,"bs--")               # more line chats in single plot
plt.show()                                        # "ro-" is a shorthand for red color, circle marker and solid line. "bs--" is a shorthand for blue color, square marker and dashed line.
```

# Subplots

```python
x = np.arange(1,20,2)
y = np.square(x)
z = np.sqrt(x)
p = np.log(x)

plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("Square of x")

plt.subplot(2,2,2)
plt.plot(x,z)
plt.title("Square root of x")

plt.subplot(2,2,3)
plt.plot(x,p)
plt.title("Log of x")
```

```python
x = np.linspace(0,10,100)

fig,axes = plt.subplots(3,3)

axes[0,0].plot(x,np.sin(x))
axes[0,0].set_title('sin fun')
axes[0,0].set_xlabel("x")
axes[0,0].set_ylabel("y")

axes[0,1].plot(x,np.cos(x))
axes[0,1].set_title('cos fun')
axes[0,1].set_xlabel("x")
axes[0,1].set_ylabel("y")

axes[0,2].plot(x,np.tan(x))
axes[0,2].set_title('tan fun')
axes[0,2].set_xlabel("x")
axes[0,2].set_ylabel("y")

axes[1,0].plot(x,np.exp(x))
axes[1,0].set_title('cos fun')
axes[1,0].set_xlabel("x")
axes[1,0].set_ylabel("y")

axes[1,1].plot(x,np.sqrt(x))
axes[1,1].set_title('sqrt')
axes[1,1].set_xlabel("x")
axes[1,1].set_ylabel("y")

axes[1,2].plot(x,np.square(x))
axes[1,2].set_title('square')
axes[1,2].set_xlabel("x")
axes[1,2].set_ylabel("y")

plt.tight_layout()
```

# Scatter plot

```python
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10

plt.scatter(x,y,marker = "o",edgecolor="green",linewidth=2,alpha=0.4,s=40,label="random")
plt.title("X vs Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend("upper left")
plt.show()
```

# Bar plot

```python
x = [1,2,3,4,5]
students = [33,45,25,30,40]
hatches=["/","+","o","*","-"]
colors=["blue","green","yellow","orange","red"]

plt.bar(x,students,color=colors,edgecolor="black",width=0.5,align="edge",linewidth=5,hatch=hatches)
plt.xlabel("Class")
plt.ylabel("no of students")
plt.title("Students in each class",fontdict={"fontsize":16,"fontweight":10,"fontname":"arial","color":"green"})
plt.xticks([1,2,3,4,5],["A","B","C","D","E"])
plt.yticks([10,80])
plt.show()
```

# Hist plot

```python
age = [23,53,56,23,76,46,98,65,35,74,32]
hatches=["/","+","o","*","-"]

sorted_age = np.sort(age)
plt.hist(sorted_age,bins=5,orientation="horizontal",linewidth=4,edgecolor="black",color='gold',hatch=hatches)
plt.xlabel("frequency")
plt.ylabel("ages")
plt.show()
```

# Pie chart

```python
x = [45,55,65,76,32]
y = ["math","eng","sci","hin","tel"]
explode=[0.2,0.0,0.0,0.0,0.0]
colors=["blue","green","yellow","orange","red"]

plt.pie(x,labels=y,shadow=True,explode=explode,colors=colors,autopct="%.2f")
plt.legend("upper left")
plt.show()
```

# 3D Scattter plot

```python
ax = plt.axes(projection="3d")

x = np.random.rand(100) *10
y = np.random.rand(100) *10
z = np.random.rand(100) *10

ax.scatter(x,y,z)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("3D scatter plot")
plt.show()
```

# 3D Line plot

```python
ax = plt.axes(projection="3d")

x = np.arange(0,50,0.1)
y = np.sin(x)
z = np.cos(x)

ax.plot(x,y,z)
plt.show()
```

# 3D Bar plot

```python
ax = plt.axes(projection="3d")

x = np.random.randint(1,10,5) * 0.5
y = np.random.randint(1,10,5)
z = np.zeros(5)

dx = np.ones(5) * 0.5
dy = np.ones(5)
dz = np.random.randint(1,10,5)
print(x,y,z,dx,dy,dz)

ax.bar3d(x,y,z,dx,dy,dz)
plt.show()
```

# 3D Plot surface

```python
ax = plt.axes(projection="3d")

x = np.arange(-5,5,0.25)
y = np.arange(-5,5,0.25)
x,y = np.meshgrid(x,y)
r = np.sqrt(x**2 + y**2)
z = np.sin(r)

ax.plot_surface(x,y,z,cmap="magma")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1000,1500,1800,1200,1600]

plt.plot(x,y,linewidth=2,linestyle='--',marker='+',label='price',color='b')
plt.xlabel('month')
plt.ylabel('sales')
plt.title('MOnthly sales')
plt.legend(loc='upper left',fontsize=10)
plt.grid(color='gray',linewidth=2,linestyle=':')
plt.xlim(0,6)
plt.ylim(500,2000)
plt.xticks([1,2,3,4,5],['jan','feb','mar','apr','may'])
plt.show()
```

```python
regions=['North','south','east','west']
gdp = [3000,5500,4000,2700]

plt.pie(gdp,labels=regions,autopct='%1.2f%%',colors=['b','g','y','r'])
plt.title("gdp of each region")
plt.show()
```
