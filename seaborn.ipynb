# Count plot

```python
data = sns.load_dataset("tips").head(20)
data
```

```python
# palette="bwr"
sns.countplot(x='sex',data=data,hue='smoker',palette='dark:y',saturation=0.4)
plt.show()
```

# Violine plot

```python
d = sns.load_dataset("tips")
d
```

```python
# order=["Lunch","Dinner"] color='r'
sns.violinplot(x="day",y="total_bill",data=d,hue="time",linewidth=2,palette="cool",saturation=0.6,)
plt.show()
```

```python
# density_norm='count','area','width'
# inner='quart','box','quartile','point','stick',None
sns.violinplot(x="time",y="total_bill",data=d,linewidth=2,palette="cool",saturation=0.6,hue='sex',density_norm='count',inner='point')
plt.show()
```

```python
sns.violinplot(x=d["total_bill"])
plt.show()
```

# Pair plot

```python
p = sns.load_dataset("tips")
p
```

```python
sns.pairplot(p,hue='sex',hue_order=['Female','Male'],palette='plasma',markers=['o','^'])
plt.show()
```

```python
sns.pairplot(p,hue='sex',hue_order=['Female','Male'],palette='plasma',x_vars=['time','sex'],y_vars=['total_bill','tip'])
plt.show()
```

```python
# kind='reg','scatter','kde','hist'
# diag_kind='kde','hist'
sns.pairplot(p,vars=['total_bill','tip'],hue='sex',hue_order=['Female','Male'],palette='plasma',kind='kde',diag_kind='kde')
plt.show()
```

# Strip plot

```python
s = sns.load_dataset("tips")
s
```

```python
# jitter=1, marker=['o','^']
sns.stripplot(x='day',y='total_bill',data=s,hue='sex',palette='rocket',linewidth=0.5,edgecolor='b',size=5,alpha=0.6)
plt.show()
```

```python
sns.stripplot(x=s['total_bill'])
plt.show()
```

# Box plot

```python
b = sns.load_dataset('tips')
b
```

```python
sns.set(style='whitegrid')
order1 = ['Fri','sun','Sat','Thur']
sns.boxplot(x='day',y='total_bill',data=b,hue='sex',linewidth=2,palette='rainbow',order=order1)
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=b,linewidth=2,color='g',showmeans=True,meanprops={'marker':'+','markeredgecolor':'b','markersize':10})
plt.show()
```

```python
sns.boxplot(x='total_bill',data=b)
plt.show()
```

# Cat plot

```python
f = sns.load_dataset('tips')
f
```

```python
# kind = 'strip','swarm','box',violin','boxen','point','bar','count'
sns.catplot(x='day',y='tip',data=f,kind='boxen',hue='sex',palette='cool',height=5,)
plt.show()
```

```python
sns.catplot(x='day',y='tip',data=f,kind='swarm')
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',data=f)
sns.set(style='whitegrid')
sns.despine()
plt.figure(figsize=(12,5))
plt.show()
```
